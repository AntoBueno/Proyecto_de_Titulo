```python
from skimage import data
from skimage import io
from matplotlib import pyplot as plt
from skimage import color
```

```python
halo_image = plt.imread('primerhalo.png') 
type(halo_image)
```

```python
def show_image(image, title='Image', cmap_type='gray'): 
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
halo_gris = color.rgb2gray(halo_image)
show_image(halo_gris)
```

# Algoritmo de SOBEL

```python
def plot_comparison(original, filtered, title_filtered):
    fig, (ax1, ax2) = plt.subplots(ncols=2, 
                                   figsize=(15, 13), 
                                   sharex=True, 
                                   sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray) 
    ax1.set_title('original') 
    ax1.axis('off')
    ax2.imshow(filtered, cmap=plt.cm.gray) 
    ax2.set_title(title_filtered) 
    ax2.axis('off')
```

```python
from skimage.filters import sobel 
# Aplico filtro sobel detecci√≥n de bordes
edge_sobel = sobel(astro_gris)
# Comparo
plot_comparison(astro_gris, edge_sobel, 
                "Edge with Sobel")
```

# Suavizado

```python
from skimage.filters import gaussian 
# Aplico el filtro
gaussian_image = gaussian(halo_gris, multichannel=True)
# Comparo
plot_comparison(halo_gris, gaussian_image, "Blurred with Gaussian filter")
```

# CONTRASTE

## Ecualizaci√≥n de histograma

```python
from skimage import exposure
# Obtener la imagen ecualizada
image_eq = exposure.equalize_hist(halo_gris)
# Show original and result
show_image(halo_gris, 'Original') 
show_image(image_eq, 'Histogram equalized')
  
```

## CLAHE

```python
from skimage import exposure 
# Apply adaptive Equalization
image_adapteq = exposure.equalize_adapthist(halo_gris, clip_limit=0.03)
# Show original and result
show_image(halo_gris, 'Original') 
show_image(image_adapteq, 'Adaptive equalized')
```

```python
plot_comparison(image_eq, image_adapteq, "Edge with Sobel")
```

```python
plt.title('Histogram of image')
plt.hist(halo_gris.ravel(), bins=256)
plt.show()
```

# Transformaciones
## Rotaci√≥n sentido reloj

```python
from skimage.transform import rotate
# Rotar la imagen 90 grados sentido reloj
image_rotated = rotate(halo_gris, -90)
# mostrar
show_image(halo_gris, 'Original') 
show_image(image_rotated, 'Rotated 90 degrees clockwise')
```

## Rotaci√≥n contra-sentido reloj

```python
from skimage.transform import rotate
# Rotar la imagen 90 grados antireloj
image_rotated = rotate(halo_gris, 90)
show_image(halo_gris, 'Original')
show_image(image_rotated, 'Rotated 90 degrees anticlockwise')
```

## Rescalar una imagen

```python
from skimage.transform import rescale
# Volver a escalar la imagen para ser 4 veces m√°s peque√±a
image_rescaled = rescale(halo_gris, 1/4, anti_aliasing=True, multichannel=True)
show_image(halo_gris, 'Original image') 
show_image(image_rescaled, 'Rescaled image')
```

## Redimensionando

```python
from skimage.transform import resize
# tama√±o de altura y ancho
height = 400 
width = 500
# Reescalo la imagen
image_resized = resize(halo_gris, (height, width), anti_aliasing=True)
# Mostrar
show_image(halo_gris, 'Original image') 
show_image(image_resized, 'Resized image')
```

```python
# Import the module and function
from skimage.transform import resize

# Set proportional height and width so it is half its size
height = int(halo_gris.shape[0] / 2.0)
width = int(halo_gris.shape[1] / 2.0)

# Resize using the calculated proportional height and width
image_resized = resize(halo_gris, (height, width),
                       anti_aliasing=True)

# Show the original and rotated image
show_image(halo_gris, 'Original')
show_image(image_resized, 'Resized image')
```
